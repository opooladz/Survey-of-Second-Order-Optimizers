```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
train_loss_sgd = np.load('./loss_acc_timing/mnist_cnn_SGD_train_loss.npy')
test_acc_sgd = np.load('./loss_acc_timing/mnist_cnn_SGD_test_acc.npy')
train_loss_adam = np.load('./loss_acc_timing/mnist_cnn_Adam_train_loss.npy')
test_acc_adam = np.load('./loss_acc_timing/mnist_cnn_Adam_test_acc.npy')
train_loss_lm = np.load('./loss_acc_timing/mnist_cnn_LM_train_loss.npy')
test_acc_lm = np.load('./loss_acc_timing/mnist_cnn_LM_test_acc.npy')
```

```python
fig, (ax1, ax2) = plt.subplots(2,1, sharex=True, figsize=(6, 8))
ax1.plot(train_loss_sgd, label='SGD')
ax1.plot(train_loss_adam, label='Adam')
ax1.plot(train_loss_lm, label='LM')
ax1.set_title('training loss curve')
ax1.legend()
ax2.plot(test_acc_sgd, label='SGD')
ax2.plot(test_acc_adam, label='Adam')
ax2.plot(test_acc_lm, label='LM')
ax2.set_title('testing acc curve')
ax2.set_ylim([0, 1])
ax2.set_xlabel('iter')
ax2.legend()
```
