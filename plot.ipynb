```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
train_loss_sgd = np.load('./loss_acc_timing/mnist_cnn_SGD_train_loss.npy')
train_acc_sgd = np.load('./loss_acc_timing/mnist_cnn_SGD_train_acc.npy')
test_loss_sgd = np.load('./loss_acc_timing/mnist_cnn_SGD_test_loss.npy')
test_acc_sgd = np.load('./loss_acc_timing/mnist_cnn_SGD_test_acc.npy')

train_loss_adam = np.load('./loss_acc_timing/mnist_cnn_Adam_train_loss.npy')
train_acc_adam = np.load('./loss_acc_timing/mnist_cnn_Adam_train_acc.npy')
test_loss_adam = np.load('./loss_acc_timing/mnist_cnn_Adam_train_loss.npy')
test_acc_adam = np.load('./loss_acc_timing/mnist_cnn_Adam_test_acc.npy')

train_loss_lm = np.load('./loss_acc_timing/mnist_cnn_LM_train_loss.npy', allow_pickle=True)
train_acc_lm = np.load('./loss_acc_timing/mnist_cnn_LM_train_acc.npy', allow_pickle=True)
test_loss_lm = np.load('./loss_acc_timing/mnist_cnn_LM_test_loss.npy', allow_pickle=True) 
test_acc_lm = np.load('./loss_acc_timing/mnist_cnn_LM_test_acc.npy', allow_pickle=True)
```

```python
fig, axs = plt.subplots(2,2, sharex=True, figsize=(14, 10))
axs[0,0].plot(train_loss_sgd, label='SGD')
axs[0,0].plot(train_loss_adam, label='Adam')
axs[0,0].plot(train_loss_lm[:50], label='LM')
axs[0,0].set_title('training loss curve')
axs[0,0].legend()

axs[0,1].plot(test_loss_sgd, label='SGD')
axs[0,1].plot(test_loss_adam, label='Adam')
axs[0,1].plot(test_loss_lm[:50], label='LM')
axs[0,1].set_title('testing loss curve')
axs[0,1].legend()

axs[1,0].plot(train_acc_sgd, label='SGD')
axs[1,0].plot(train_acc_adam, label='Adam')
axs[1,0].plot(train_acc_lm[:50], label='LM')
axs[1,0].set_title('training acc curve')
axs[1,0].set_ylim([0, 1])
axs[1,0].set_xlabel('iter')
axs[1,0].legend()


axs[1,1].plot(test_acc_sgd, label='SGD')
axs[1,1].plot(test_acc_adam, label='Adam')
axs[1,1].plot(test_acc_lm[:50], label='LM')
axs[1,1].set_title('testing acc curve')
axs[1,1].set_ylim([0, 1])
axs[1,1].set_xlabel('iter')
axs[1,1].legend()
```
