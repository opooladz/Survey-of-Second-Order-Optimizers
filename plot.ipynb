```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
train_loss_sgd = np.load('./loss_acc_timing/mnist_cnn_SGD_train_loss.npy')
test_acc_sgd = np.load('./loss_acc_timing/mnist_cnn_SGD_test_acc.npy')
train_loss_adam = np.load('./loss_acc_timing/mnist_cnn_Adam_train_loss.npy')
test_acc_adam = np.load('./loss_acc_timing/mnist_cnn_Adam_test_acc.npy')
train_loss_lm = np.load('./loss_acc_timing/mnist_cnn_LM_train_loss.npy')
test_acc_lm = np.load('./loss_acc_timing/mnist_cnn_LM_test_acc.npy')
train_loss_lbfgs = np.load('./loss_acc_timing/mnist_cnn_lbfgs_train_loss.npy',allow_pickle=True)
test_acc_lbfgs = np.load('./loss_acc_timing/mnist_cnn_lbfgs_test_acc.npy',allow_pickle=True)
train_loss_hf = np.load('./loss_acc_timing/mnist_cnn_HF_train_loss.npy',allow_pickle=True)
test_acc_hf = np.load('./loss_acc_timing/mnist_cnn_HF_test_acc.npy',allow_pickle=True)
train_loss_ekfac = np.load('./loss_acc_timing/mnist_cnn_EKFAC_train_loss.npy',allow_pickle=True)
test_acc_ekfac = np.load('./loss_acc_timing/mnist_cnn_EKFAC_test_acc.npy',allow_pickle=True)
train_loss_kfac = np.load('./loss_acc_timing/mnist_cnn_KFAC_train_loss.npy',allow_pickle=True)
test_acc_kfac = np.load('./loss_acc_timing/mnist_cnn_KFAC_test_acc.npy',allow_pickle=True)
train_loss_ekfac_adam = np.load('./loss_acc_timing/mnist_cnn_EKFAC-Adam_train_loss.npy',allow_pickle=True)
test_acc_ekfac_adam = np.load('./loss_acc_timing/mnist_cnn_EKFAC-Adam_test_acc.npy',allow_pickle=True)
train_loss_kfac_adam = np.load('./loss_acc_timing/mnist_cnn_KFAC-Adam_train_loss.npy',allow_pickle=True)
test_acc_kfac_adam = np.load('./loss_acc_timing/mnist_cnn_KFAC-Adam_test_acc.npy',allow_pickle=True)
```

```python
fig, (ax1, ax2) = plt.subplots(2,1, sharex=True, figsize=(6, 8))
ax1.plot(train_loss_sgd, label='SGD')
ax1.plot(train_loss_adam, label='Adam')
ax1.plot(train_loss_lm, label='LM')
ax1.plot(train_loss_lbfgs, label='LBFGS')
ax1.plot(train_loss_hf, label='HF')
ax1.plot(train_loss_ekfac, label='EKFAC')
ax1.plot(train_loss_kfac, label='KFAC')
ax1.plot(train_loss_ekfac_adam, label='EKFAC-Adam')
ax1.plot(train_loss_kfac_adam, label='KFAC-Adam')
ax1.set_title('training loss curve')
ax1.legend()
ax2.plot(test_acc_sgd, label='SGD')
ax2.plot(test_acc_adam, label='Adam')
ax2.plot(test_acc_lm, label='LM')
ax2.plot(test_acc_lbfgs, label='lbfgs')
ax2.plot(test_acc_hf, label='HF')
ax2.plot(test_acc_ekfac, label='EKFAC')
ax2.plot(test_acc_kfac, label='KFAC')
ax2.plot(test_acc_ekfac_adam, label='EKFAC-Adam')
ax2.plot(test_acc_kfac_adam, label='KFAC-Adam')
ax2.set_title('testing acc curve')
ax2.set_ylim([0, 1])
ax2.set_xlabel('iter')
ax2.legend()
```
